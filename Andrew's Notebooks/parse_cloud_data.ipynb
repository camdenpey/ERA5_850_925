```python
months_list = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
year = 1990
starting_directory = "S:/Science_and_Research/ERA5_Reanalysis_1990-2020/VIH/"
```

```python
data = pd.read_csv(starting_directory + "1990/Vichy_1990_January_clouds.csv")
```

```python
cloud_sum = []
month_counter = 0
end_sum = 0
```

```python
for i in range(31):
    for k in range(12):
        for j in range(round(len(data)/13)):
            if j == len(data)/13-1:
                end_sum += sum(data["tcc"][1+j*13:])
                if month_counter != 11:
                    month_counter += 1
                    data =pd.read_csv(starting_directory+str(year)+"/Vichy_"+str(year)+"_"+months_list[month_counter]+"_clouds.csv")
                    end_sum += data["tcc"][0]
                    end_sum = end_sum/13
                    cloud_sum.append(end_sum)
                else:
                    if year < 2020:
                        year += 1
                        month_counter = 0
                        data = pd.read_csv(starting_directory+str(year)+"/Vichy_"+str(year)+"_"+months_list[month_counter]+"_clouds.csv")
                        print(starting_directory+str(year)+"/Vichy_"+str(year)+"_"+months_list[month_counter]+"_clouds.csv")
                        end_sum += data["tcc"][0]
                        end_sum = end_sum/13
                        cloud_sum.append(end_sum)
                    else:
                        cloud_sum.append(end_sum/12)
                        break
            else:
                cloud_sum.append(sum(data["tcc"][1+j*13:13 + j*13+1])/13)
```

```python
for i in range(len(cloud_sum)):
    if cloud_sum[i] < 6/8:
        cloud_sum[i] = 0
    else:
        cloud_sum[i] = 1
```
